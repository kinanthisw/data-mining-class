## Imputation Missing Value

```python
np.sum(data.isnull()) #number of NA for each variable
```

```python
data.isnull().sum()
```

```python
data.isnull().sum().sum()
```

```python
#imputasi missing value data numerik
data['CAMKII_N'] = data['CAMKII_N'].fillna((data['CAMKII_N'].mean())) 
data['CREB_N'] = data['CREB_N'].fillna((data['CREB_N'].mean()))
data['ELK_N'] = data['ELK_N'].fillna((data['ELK_N'].median()))
data['ERK_N'] = data['ERK_N'].fillna((data['ERK_N'].mean()))
data['GSK3B_N'] = data['GSK3B_N'].fillna((data['GSK3B_N'].mean()))
data['JNK_N'] = data['JNK_N'].fillna((data['JNK_N'].mean()))
data['MEK_N'] = data['MEK_N'].fillna((data['MEK_N'].mean()))
data['TRKA_N'] = data['TRKA_N'].fillna((data['TRKA_N'].mean()))
data['RSK_N'] = data['RSK_N'].fillna((data['RSK_N'].mean()))
data.isnull().any()
```

```python
# imputasi untuk data numerik yang lain
data['APP_N'] = data['APP_N'].fillna((data['APP_N'].mean())) 
data['Bcatenin_N'] = data['Bcatenin_N'].fillna((data['Bcatenin_N'].median()))
data['SOD1_N'] = data['SOD1_N'].fillna((data['SOD1_N'].median()))
data['DSCR1_N'] = data['DSCR1_N'].fillna((data['DSCR1_N'].mean()))
data['BAD_N'] = data['BAD_N'].fillna((data['BAD_N'].mean()))
data['BCL2_N'] = data['BCL2_N'].fillna((data['BCL2_N'].mean()))
data['pCFOS_N'] = data['pCFOS_N'].fillna((data['pCFOS_N'].mean()))
data['H3AcK18_N'] = data['H3AcK18_N'].fillna((data['H3AcK18_N'].mean()))
data['EGR1_N'] = data['EGR1_N'].fillna((data['EGR1_N'].mean()))
data['H3MeK4_N'] = data['H3MeK4_N'].fillna((data['H3MeK4_N'].mean()))
data.isnull().any()
```

```python
#imputasi data kategorik dengan modus
data = data.apply(lambda x:x.fillna(x.value_counts().index[0]))
data.isnull().any()
```

## Encode Labels

```python
from sklearn.preprocessing import LabelEncoder
```

```python
# Encode untuk Sex, Steroid, Anaroxia, Spleen Pabable, Spiders tanpa membuat kolom baru
data['Genotype'] = LabelEncoder().fit_transform(data['Genotype'])
data['Treatment'] = LabelEncoder().fit_transform(data['Treatment'])
data['Behavior'] = LabelEncoder().fit_transform(data['Behavior'])
data['class'] = LabelEncoder().fit_transform(data['class'])
data.head()
```
